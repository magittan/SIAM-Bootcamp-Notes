# Log Returns

When performining analysis on financial data, people typically use daily return instead of prices. This can be done with log returns since it approximates the daily return.
- Log Returns have a mean at zero typically
- Prices typically are moving up and are also all starting at arbitrary points (50 Dollars, 250 Dollars, etc...)

```python
plt.hist(data[0], bins=30)
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Histogram of Log Returns")
```

```python
#Determination of returns w.r.t log returns, this is very similar to simple returns
def derive_log_returns(price_data,daily_rf_rate=0):
    return np.diff(np.log(price_data), axis=1) - daily_rf_rate
```

```python
log_data = derive_log_returns(data)
```

```python
log_data.shape
```

```python
plt.hist(log_data[0], bins=30)
plt.xlabel("Log Returns")
plt.ylabel("Frequency")
plt.title("Histogram of Log Returns")
```

# Determining Covariance Matrix

```python
def determine_covariance(time_period_start,time_period_end,returns):
    return np.cov(returns[time_period_start:time_period_end,:].T)
```

```python
covariance_matrix = np.cov(log_data)
```

```python
plt.imshow(covariance_matrix)
plt.colorbar()
```

```python
# Fancier Way of Doing this Estimation
estimator = LedoitWolf().fit(log_data.T)
plt.imshow(estimator.covariance_)
plt.colorbar()
```

# Markowitz

```python
# Estimating the Covariance
estimator = LedoitWolf().fit(log_data.T)
```

```python
MVO = M.Markowitz(log_data.T.sum(axis=0),estimator.covariance_)
calculated_weights = MVO.design_portfolio_for_return(0.8)
```

```python
calculated_weights.sum()
```

```python
calculated_weights.shape
```

```python
plt.title("Markowitz Portfolio")
plt.plot(np.dot(calculated_weights,log_data).cumsum())
plt.ylabel("Return")
plt.xlabel("Time (Days)")
```

```python
plt.title("Uniform Portfolio")
weights = np.ones(16)/16
plt.plot(np.dot(weights,log_data).cumsum())
plt.ylabel("Return")
plt.xlabel("Time (Days)")
```
