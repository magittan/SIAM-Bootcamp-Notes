```python
# General Purpose
import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import odeint

%matplotlib inline

# Jupyter Specifics
from IPython.display import HTML
from ipywidgets.widgets import interact, IntSlider, FloatSlider, Layout

style = {'description_width': '180px'}
slider_layout = Layout(width='95%')
```

```python
def main(fish_birthrate, fish_deathrate, bear_birthrate, bear_deathrate, initial_fishes, initial_bears, days):

    def function(s, t):
        x, y = s
        dydt = [
            fish_birthrate * x - fish_deathrate * x * y,
            bear_birthrate * x * y - bear_deathrate * y
        ]

        return dydt

    time = np.arange(0, days, 0.01)
    initial_conditions = [initial_fishes, initial_bears]
    solution = odeint(function, initial_conditions, time)

    #Graphic details
    fig, axes = plt.subplots(1, 2, figsize=(15, 10))

    ax = axes[0]

    ax.plot(time, solution[:, 0], label='Fishes(t)')
    ax.plot(time, solution[:, 1], label='Bears(t)')

    if days <= 30:
        step = 1
        rotation = "horizontal"
    elif days <= 150:
        step = 5
        rotation = "vertical"
    else:
        step = 10
        rotation = "vertical"

    ax.set_xticklabels(np.arange(0, days + 1, step, dtype=np.int), rotation=rotation)
    ax.set_xticks(np.arange(0, days + 1, step))

    ax.set_xlim([0, days])
    ax.set_ylim([0, max(max(solution[:, 0]), max(solution[:, 1])) * 1.05])
    ax.set_xlabel('Time')
    ax.set_ylabel('Population')
    ax.legend(loc='best')
    ax.grid()


    ax = axes[1]

    ax.plot(solution[:, 0], solution[:, 1], label='Bears vs Fishes')

    ax.set_xlim([0, max(solution[:, 0]) * 1.05])
    ax.set_ylim([0, max(solution[:, 1]) * 1.05])
    ax.set_xlabel('Fishes')
    ax.set_ylabel('Bears')
    ax.legend(loc='best')
    ax.grid()

    plt.tight_layout()
    plt.show()

interact(main, fish_birthrate=FloatSlider(min=0, max=24, step=0.01, value=1, description='Birth Rate of Fishes', style=style, layout=slider_layout),
               fish_deathrate=FloatSlider(min=0, max=24, step=0.01, value=1, description='Death Rate of Fishes', style=style, layout=slider_layout),
               bear_birthrate=FloatSlider(min=0, max=24, step=0.01, value=1, description='Birth Rate of Bears', style=style, layout=slider_layout),
               bear_deathrate=FloatSlider(min=0, max=24, step=0.01, value=1, description='Death Rate of Bears', style=style, layout=slider_layout),
               initial_fishes=FloatSlider(min=0 , max=100, step=1, value=2, description='Initial Fishes', style=style, layout=slider_layout),
               initial_bears=FloatSlider(min=0 , max=100, step=1, value=1, description='Initial Bears', style=style, layout=slider_layout),
               days=FloatSlider(min=0 ,max=365 , step=10, value=15, description='Total number of Days', style=style, layout=slider_layout),
        );
```

```python
import numpy as np
import scipy.integrate as spi
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
m = 1.  # particle's mass
k = 1.  # drag coefficient
g = 9.81  # gravity acceleration
```

```python
# The initial position is (0, 0).
v0 = np.zeros(4)
# The initial speed vector is oriented
# to the top right.
v0[2] = 4
v0[3] = 10
```

```python
def f(v, t0, k):
    # v has four components: v=[u, u'].
    u, udot = v[:2], v[2:]
    # We compute the second derivative u'' of u.
    udotdot = -k / m * udot
    udotdot[1] -= g
    # We return v'=[u', u''].
    return np.r_[udot, udotdot]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 4))

# We want to evaluate the system on 30 linearly
# spaced times between t=0 and t=3.
t = np.linspace(0., 3., 30)

# We simulate the system for different values of k.
for k in np.linspace(0., 1., 5):
    # We simulate the system and evaluate $v$ on the
    # given times.
    v = spi.odeint(f, v0, t, args=(k,))
    # We plot the particle's trajectory.
    ax.plot(v[:, 0], v[:, 1], 'o-', mew=1, ms=8,
            mec='w', label=f'k={k:.1f}')
ax.legend()
ax.set_xlim(0, 12)
ax.set_ylim(0, 6)
```

Order, linearity

In some situation the rate of change in the number of individuals in a population at a specific time is directly proportional to the number of individuals at that specific time: $$ \frac{dP}{dt} \propto P $$ where $ P $ is the population at time $ t $.
By introducting a constant of proportionality we have: $$ \frac{dP}{dt} = kP $$ which we can solve.
$$ \frac { 1 }{ P } dP=kdt\\ \int { \frac { 1 }{ P }  } dP=\int { k } dt\\ \ln { \left| P \right|  } =k{ t }+c\\ \ln { \left| P \right|  } =k{ t }+c\\ P\left( t \right) ={ e }^{ k{ t }+c }={ e }^{ c }{ e }^{ k{ t } }\\ P\left( t \right) =c{ e }^{ k{ t } } $$
If $ P \left( 0 \right) = {P}_{0} $ we have: $$ P\left( 0 \right) =c{ e }^{ \left( 0 \right)  }=c={ P }_{ 0 }\\ P\left( t \right) ={ P }_{ 0 }{ e }^{ k{ t } } $$
