## What is Graph Theory?
<b> Graph Theory </b> is the mathematical study of the properties of graphs and their applications.

A graph is comprised of <b> vertices </b> (also referred to as nodes) and <b> edges </b> that join them. 

Two vertices are <b> adjacent </b> if they are connected by an edge. 

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline

n = 10  # Number of nodes in the graph.
# Each node is connected to the two next nodes,
# in a circular fashion.
adj = [(i, (i + 1) % n) for i in range(n)]
adj += [(i, (i + 2) % n) for i in range(n)]

g = nx.Graph(adj)

fig, ax = plt.subplots(1, 1, figsize=(6, 6))
nx.draw_circular(g, ax=ax)
```

```python
n = 5
g = nx.Graph()
edges = [(1,2), (2,3), (3,4), (4,5), (5,1)]
g.add_edges_from(edges)
nx.draw_circular(g, with_labels = True)
```

In this case, vertices 1 and 2 are adjacent, but vertices 1 and 3 would be considered <b> nonadjacent </b>. 

## Representing Graphs
A graph G is denoted by a <b> set of vertices, V </b> and a  <b> set of edges, E</b>. This is written as G = (V, E).

```python
n = 4
g = nx.Graph()
edges = [(1,2), (1,3), (2,3), (3,4), (4,1)]
g.add_edges_from(edges)
nx.draw(g, with_labels = True)
```

For the graph G above, the vertices can be described as the set [1, 2, 3, 4] and the edges can be described as the [12, 13, 14, 23, 34] for a total of 4 vertices and 5 edges.

## Complete Graphs
What is the most number of edges a graph on n vertices can have, provided that it is a <b> simple graph?</b> 

Connecting each vertice to every other vertice in the graph answers this question. A graph in which every pair of vertices is connected, such as the one below, is known as a <b> complete graph </b>

```python
n=8


g = nx.Graph()
e_list = []

for i in range (n):
    for j in range (n):
        e_list += [(i,j)]

g.add_edges_from(e_list)
nx.draw_circular(g, with_labels = True)
```

Try adjusting the number of edges in the <b> complete graph </b> below by changing the value of n.

```python
n=5
g = nx.Graph()
e_list = []

for i in range (n):
    for j in range (n):
        e_list += [(i,j)]

g.add_edges_from(e_list)
nx.draw_circular(g, with_labels = True)
```

## Complementary Graphs
Every graph has a complement. 
<br>
<br> A <b> complementary graph </b> contains the same number of vertices (V) as the original graph, but the opposite edges. Every connected pair of vertices in the original graph is disconnected in the complement, and every disconnected pair in the original graph is connected in the complement. 
<br>
<br> Here is an example of a pair of complementary graphs:

```python
n = 5
g = nx.Graph()
## Original Graph
g.add_edge(1,2,color='b',weight=1)
g.add_edge(1,3,color='b',weight=1)
g.add_edge(2,3,color='b',weight=1)
g.add_edge(3,4,color='b',weight=1)
g.add_edge(4,5,color='b',weight=1)
g.add_edge(5,1,color='b',weight=1)

## Graph Complement
g.add_edge(1,4,color='r',weight=4)
g.add_edge(2,4,color='r',weight=4)
g.add_edge(2,5,color='r',weight=4)
g.add_edge(3,5,color='r',weight=4)

pos = nx.circular_layout(g)
edges = g.edges()
colors = [g[u][v]['color'] for u,v in edges]
weights = [g[u][v]['weight'] for u,v in edges]

nx.draw(g, pos, edges=edges, edge_color=colors, width=weights, with_labels = True)
```

In this graph, the original is drawn in blue, and the complement is drawn in red. 
<br> 
As you can see, the sum of a graph and its complement creates a complete graph.

## Example 1
<b> The Problem of the Five Princes </b>
<br> A kingdom was ruled by a king who had five sons. He wished that upon his death, the land could be divided into five continuous regions, in which every region shares a common boundary with the other four regions. Is this possible?

Try representing this situation as a graph, with 5 vertices representing each of the sons, and the edges representing shared boundaries. 

```python
n = 5
g = nx.Graph()
edges = []
g.add_edges_from(edges)
nx.draw(g, with_labels = True)
```

<b> Solution: </b>Because each territory has to share a boundary with every other territory, you must draw a complete graph. Each node should be connected to every other node in the graph. However, each son has to be given a continuous region; it is not possible to draw a complete graph on 5 vertices, in which no two edges intersect (in order for the regions to be continuous). 

## Example 2
<b> The Konigsberg Bridge Problem </b>
<br> Is it possible to walk around Konigsberg, and return to the same place you started, by crossing each of the bridges only once? 
![](https://www.physicslog.com/wp-content/uploads/2017/12/1.png)

<b> Solution: </b>Once again, this problem can be represented as a graph! Each of the land masses, [A, B, C, D], can be represented as vertices, and each of the bridges, [a, b, c, d, e, f, g], can be represented as edges. Once drawing this problem as a graph, you can see that no matter where you start on the map, you cannot cross all of the bridges and return to the same position without crossing one of the bridges twice.

## Applications of Graph Theory
<b> The Four Color Problem </b>
<br> In any map, how many colors are needed to color different regions so that every two regions sharing a common boundary are colored differently?
<br>
<br> This can be viewed as a graph problem. As shown in the first example, with <b> The Problem of the Five Princes,</b> 5 territories cannot be drawn in such a way that they all border each other. However, 4 territories can be drawn in such a way that all of them share a border. 
<br>
<br> As you can see, it is impossible to draw a complete graph of 5 vertices without some of the edges intersecting. But a complete graph of 4 vertices can be drawn without intersecting edges.

![](https://cdn.britannica.com/56/96256-050-74F047AB/vertices-paths-plane-collection-use-dimension.jpg)

Because the complete graph on 4 vertices can be drawn without any intersecting edges, but the complete graph on 5 vertices cannot be drawn in such a way, it means that in any map, there can be no more than 4 territories that all share a boundary with each other. Thus, only 4 colors are needed to color any map. 
<br>
